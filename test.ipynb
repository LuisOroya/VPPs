```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rc('font', family='serif', serif='cmr10')
plt.rcParams['text.usetex'] = False  # Disable TeX rendering
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams.update({'font.size': 14})

# Assume you have data for the following variables
T = list(df['t'].values)  # Time periods
PS = list(df['PS'].values)  # Active power values
CS = list(df2['CS'].values)  # Costs

# Setting up the figure to contain two subplots
plt.figure(figsize=(10, 4), dpi=200)

# Creating the first subplot for active power
ax1 = plt.gca()  # Get current axis
ax1.plot(T, np.array(PS), 'r', label='Active Power [kW]', drawstyle='steps-post')
ax1.set_xlabel(r'Time Period', fontsize=16)
ax1.set_ylabel(r'Active Power [kW]', fontsize=16)
# ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0), useMathText=True)
ax1.set_xlim(0.5, 96.5)
ax1.legend(loc='upper left')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-2000, 3000)

# Creating a second y-axis for costs
ax2 = ax1.twinx()
ax2.plot(T, np.array(CS), 'b--', label='CS', drawstyle='steps-post')
ax2.set_ylabel(r'Electricity Price [USD/MWh]', fontsize=16)
ax2.legend(loc='upper right')
ax2.set_ylim(40, 200)

# Auto adjust layout to avoid overlaps
plt.tight_layout()

# Saving the figure
plt.savefig('VPP Transaction.pdf', bbox_inches='tight', pad_inches=0.02)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('EV_data.csv', names=['EV', 'Time', 'Discharge', 'Charge', 'SOC'])

# Define a list of colors to cycle through
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange']  # Expand this list if more than 8 EVs

# Plot Discharge and Charge Power for all EVs on the same plot
plt.figure(figsize=(10, 6), dpi=200)
for ev in data['EV'].unique():
    ev_data = data[data['EV'] == ev]
    color = colors[(ev-1) % len(colors)]  # Repeat colors if more EVs than colors
    
    # Plotting discharge as negative
    plt.plot(ev_data['Time'], -ev_data['Discharge'], label=f'EV {ev}', linestyle='-', color=color,drawstyle='steps-post')
    # Plotting charge as positive
    plt.plot(ev_data['Time'], ev_data['Charge'], linestyle='-', color=color,drawstyle='steps-post')

plt.title('Combined Discharge (Negative) and Charge (Positive) Power for All EVs')
plt.xlabel('Time')
plt.ylim(-40, 60)
plt.ylabel('Power (kW)')
plt.legend()
plt.show()

# Plot for SOC for all EVs
plt.figure(figsize=(10, 6), dpi=200)
for ev in data['EV'].unique():
    ev_data = data[data['EV'] == ev]
    plt.plot(ev_data['Time'], ev_data['SOC'], label=f'EV {ev}', linestyle='--',drawstyle='steps-post')
plt.title('SOC for All EVs')
plt.xlabel('Time')
plt.ylabel('SOC (kW)')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('tradepq.csv')

# Extract unique weights
weights = data['w1'].unique()

# Define the prosumers and time steps
prosumers = [17, 23, 30]
time_steps = range(1, 97)  # Assuming 96 values per weight

# Plot Active Power (PexPRO) for each prosumer
for prosumer in prosumers:
    plt.figure(figsize=(10, 4), dpi=200)
    for w in weights:
        subset = data[data['w1'] == w]
        plt.plot(time_steps, subset[f'PexPRO_{prosumer}'].values, linestyle='--', label=f'w={w:.2f}', linewidth=1.0)
    
    # Calculate and plot the average curve
    avg_PexPRO = data.groupby(data.index % 96)[f'PexPRO_{prosumer}'].mean()
    plt.plot(time_steps, avg_PexPRO, linestyle='-', color='black', label='Average', linewidth=1)
    
    plt.title(f'Active Power Exchange (PexPRO) for Prosumer {prosumer}')
    plt.xlabel('Time Step')
    plt.ylabel('Active Power (kW)')
    plt.legend(loc='center right', fontsize=10)
    plt.xlim(0, 120)
    plt.grid(True)
    plt.savefig(f'PexPRO_{prosumer}.png', dpi=300, bbox_inches='tight')
    plt.show()

# Plot Reactive Power (QexPRO) for each prosumer
for prosumer in prosumers:
    plt.figure(figsize=(10, 4), dpi=200)
    for w in weights:
        subset = data[data['w1'] == w]
        plt.plot(time_steps, subset[f'QexPRO_{prosumer}'].values, linestyle='--', label=f'w={w:.2f}', linewidth=1.0)
    
    # Calculate and plot the average curve
    avg_QexPRO = data.groupby(data.index % 96)[f'QexPRO_{prosumer}'].mean()
    plt.plot(time_steps, avg_QexPRO, linestyle='-', color='black', label='Average', linewidth=1)
    
    plt.title(f'Reactive Power Exchange (QexPRO) for Prosumer {prosumer}')
    plt.xlabel('Time Step')
    plt.ylabel('Reactive Power (kW)')
    plt.legend(loc='center right', fontsize=10)
    plt.xlim(0, 120)
    plt.grid(True)
    plt.savefig(f'QexPRO_{prosumer}.png', dpi=300, bbox_inches='tight')
    plt.show()
```
